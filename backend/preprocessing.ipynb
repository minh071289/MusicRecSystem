```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
last_fm = pd.read_csv('lastfm.csv')
last_fm.head()
```

```python
last_fm.count()
```

```python
print("Max user:", last_fm['user'].max())
print("Min user:", last_fm['user'].min())
print()

print("Number of distinct users:", last_fm['user'].nunique())
print("Number of distinct artists:", last_fm['artist'].nunique())
print("Number of distinct countries:", last_fm['country'].nunique())
```

### Data Cleaning

```python
# validate that there is no null values
cols = ['user', 'artist', 'sex', 'country']

null_counts = last_fm[cols].isnull().sum()
print(null_counts)
```

```python
# remove duplicates
last_fm1 = last_fm.drop_duplicates(subset=['user','artist','sex','country'])
print(last_fm1.shape[0])
```

```python
lastfm_df_final = (
    last_fm1
    .groupby(['user', 'sex', 'country'])['artist']
    .apply(list)  
    .reset_index()
    .rename(columns={'artist': 'artist_playlist'})
)

print(lastfm_df_final.head())
```

```python
lastfm_df_final.to_csv("lastfm_processed.csv")
```
